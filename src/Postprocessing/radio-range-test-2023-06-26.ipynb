```python
import pandas as pd
import matplotlib.pyplot as plt

# csv = pd.read_csv("src/Postprocessing/export_160m.csv")
csv = pd.read_csv("src/Postprocessing/export_270m.csv")
```

```python
import math


plt.figure()

rssi_db = [(float(str(it)[:-3]) if "db" in str(it) else math.inf) for it in csv["rssi"]]

plt.subplot(211)
plt.title("FCB V2, 270 meter test")
plt.scatter(range(len(csv["rssi"])), rssi_db, label="rssi")

plt.ylim((-60, -45))
plt.legend()
plt.subplot(212)
plt.scatter(range(len(csv["lqi"])), csv["lqi"], label="lqi")
plt.ylim((0, 20))
plt.legend()

plt.show()
```
